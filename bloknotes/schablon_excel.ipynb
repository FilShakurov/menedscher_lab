```python

from config_core.vspomogat_func import process_multiheader_column
import pandas as pd
```

```python
RENAME_COLUMNS_EXCEL_RAB = {
    'Регистрационный\n№ пробы_2': 'lab_nomer',
    'Уколы/Окатанность_6': 'ukol',
    'Наименование грунта \nпо ГОСТ 25100-2020_7': 'opisanie_razbor',
    'Природная влажность, д.е._19': 'wlashn',
    'Плотность\nг/см3_влажного\nгрунта, р_24': 'plotn',
    'Плотность частиц грунта\nр s, г/см3_26': 'udelka',
    'Содержание органического вещества, %_36': 'organika'
}

def zagr_excel(path):
    df = pd.read_excel(
        path,
        sheet_name="Сводная физ св-в",
        skiprows=8,
        header=[0, 1, 2, 3, 4],
        na_values="-"
    )

    df.columns = [process_multiheader_column(col) for col in df.columns]

    df = df.rename(columns=RENAME_COLUMNS_EXCEL_RAB)
    df = df[RENAME_COLUMNS_EXCEL_RAB.values()]
    #
    df = df.dropna(subset=['lab_nomer'])

    return df
```

```python
df = zagr_excel("../excelki/test.XLSX")
```
