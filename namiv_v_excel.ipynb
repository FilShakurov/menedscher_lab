```python
cols_1_stroka = [
    "lab_nomer",
    'kolba/naveska_first',
    'areometr_first',
    '1_zamer/temp_first',
    '2_zamer/temp_first',
    '3_zamer/temp_first',
    'gran_10_first',
    'gran_5-10_first',
    'gran_5-2_first',
    'gran_2-1_first',
    'gran_1-0,5_first',
    'gran_0,5-0,25_first',
    'gran_0,25-0,10_first'
]

cols_2_stroka = [
    "lab_nomer",
    'kolba/naveska_last',
    '1_zamer/temp_last',
    '2_zamer/temp_last',
    '3_zamer/temp_last',
]

columns_vigruzka_namiv1 = {
    "kolba/naveska_first": "kolba/naveska",
    "1_zamer/temp_first": "1_zamer/temp",
    "2_zamer/temp_first": "2_zamer/temp",
    "3_zamer/temp_first": "3_zamer/temp",
}

columns_vigruzka_namiv2={
    "kolba/naveska_last": "kolba/naveska",
    "1_zamer/temp_last": "1_zamer/temp",
    "2_zamer/temp_last": "2_zamer/temp",
    "3_zamer/temp_last": "3_zamer/temp",
}
```
